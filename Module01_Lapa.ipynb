The following code, sets up the modeling environment for this assignment.



```python
import numpy as np # Pulls in the Python library
import matplotlib.pyplot as plt # Pulls in the plotting tool from np
```

The following values are the initial parameters to the one-dimensional random walk.







```python
Nt = 200       # Number of time steps in the random walk
mu_x = 1.5     # Average step size in the random walk
s_x = 2.1      # Standard deviation of the step size in the random walk
Nx = 1000      # Minimum step size at each step in the random walk
min_x = 1.25   # Number of random walkers
```

Used AI (Copilot) to help code the while loop

```python
def generate_walk(mu_x):            # Generates a random step size from mu_x
        mu = np.random.normal(mu_x)
        while abs(mu) < min_x:        # Checks if the value is less than min_x
            walk = np.random.normal(mu_x)
        return walk                   # Returns step
```

Create a container to store the initial position of the random walker and consecutive steps each time. The container is an array of zeros that has Nt rows and Nx colummns. At each row, it corresponds to a specific time and at each column, it corresponds to a specific walker.

```python
X = np.zeros((Nt,Nx)) # Creates an array of zeros
```

Create a two for loops, one for each time step and another for each walker.

Note:

*   "x_i" calculates the random step size for the "j"th walker at the "i"th time step
*   "if" and "else" is used to check if the model is at the first time step or at a different time step
*   "if" is typically used to represent an initial position of 0
*   "else" is used to represent subsequent time steps after 0





```python
for i in np.arange(Nt):
  for j in np.arange(Nx):
      x_i = s_x*np.random.randn() + mu_x
      if (i==0):
       X[i,j]= 0.0 + x_i
      else:
        X[i,j]= X[i-1,j] + x_i
```

Plot the one-dimensional random walk

```python
mean_X = np.mean(X,axis=1) # Calculate the average distance of all walkers through time
std_X = np.std(X,axis=1) # Calculate the standard deviation of all walkers through time

plt.rcParams.update({'font.size':16}) # Font size for all text is 16

plt.figure(figsize=(16,10))
plt.plot(np.arange(Nt),X,'b-')                # Plots X data as a blue solid line
plt.plot(np.arange(Nt), mean_X,'k-')          # Plots mean_X as a black solid line
plt.plot(np.arange(Nt),mean_X + std_X, 'k--') # Plots mean_X + std_X as a black dashed line
plt.plot(np.arange(Nt),mean_X - std_X, 'k--') # Plots mean_X - std_X as a black dashed line
plt.title('One-Dimensional Random Walks')
plt.xlabel('Number of Steps')                 # x-axis label
plt.ylabel('Distance Traveled [m]')           # y-axis label
```
