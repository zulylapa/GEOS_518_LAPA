# Indirect Sources Research Project
Direct vs indirect greenhouse (GHG) emissions are classified based on the type of source. Direct emissions, also known as Scope 1 by the Environmental Protection Agency (EPA), are emissions from sources that are controlled or owned by an organization <a name="cite_ref-1"></a>[<sup>[1]</sup>](#cite_note-1). For example boilers in wastewater treatment facilities that use methane and/or gas to heat the facility or other processes. Other examples may include vehicles owned by the facility and used to transport solids to a nearby landfill. Indirect emissions, also known as Scope 2 by EPA, are emissions that are generated by purchase. For example some indirect sources can be electricity or gas (from a gas company or 3rd party)<a name="cite_ref-1"></a>[<sup>[1]</sup>](#cite_note-1). Scope 2 sources are still accounted for because they are on-site and produce GHG. There is a third scope that focuses on indirect sources that are not on-site, but are activities  that add to the GHG inventory. For example distribution, purchases of goods and services <a name="cite_ref-1"></a>[<sup>[1]</sup>](#cite_note-1).

<sup>1 </sup> [Link to the Scope 1 and 2 inventory guidance](https://www.epa.gov/climateleadership/scope-1-and-scope-2-inventory-guidance)

## Model Description
The model that will be analyzed is the Local GHG Inventory Tool: Community Module. This tool was developed by EPA to be used by states and community organizers to estimate direct and indirect emissions from on-site or off-site sources. The tool is presented in a macro-enabled excel spreadsheet.

The spreadsheet is organized in steps.


*   First, the user would need to select background data. The tool has pre-populated available data for the city, year and population, as well as the eGRID subregion, specific to the state. After selecting the necessary inputs, the spreadsheet has been set up.
*   The second step is to add available stationary data (specifically fuel type and fuel use). After input, the stationary emissions will be calculated. The next data entry is for mobile equipment, solid waste, wastewater, electricity, water, land management, and other emissions (mainly pertaining to Scope 3). Similar calculations are also run to estimate emissions.

This spreadsheet does require some level of data from the facility to best estimate the indirect emissions. The data collected by sampling can give a rough estimate of how much BOD5 is produced in kg/day, however purchased electricity information is something that would be needed from the electricity company in order to input data, or researched by online sources.

This model shares similarities with the Carbon Model (Box Model) that was overviewed earlier in the semester. In the Carbon Model, it is my understanding that there are two parameters of interest, carbon produced from the atmosphere and carbon produced from biosphere. There is also an added flux of carbon from photosynthesis that is added to the finite differential equation for carbon produced from the atmosphere. The goal of this model is to do a mass balance and use two equations to estimate carbon produced from the atmosphere and biosphere over time.

Based on how this tool is set up, the tool can calculate eight main emission sectors:


*   Stationary combustion of fossil fuels
*   Mobile combustion of fossil fields
*   Solid waste disposal
*   Wastewater treatment
*   Building/Facility Electricity Consumption
*   Water Consumption
*   Agriculture and Land Management
*   Urban Forestry
